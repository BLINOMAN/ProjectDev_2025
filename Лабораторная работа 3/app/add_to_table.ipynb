```python
from sqlalchemy import create_engine
from sqlalchemy.orm import sessionmaker
from orm.main import User, Address, Base

# Подключение к PostgreSQL
engine = create_engine(
    "postgresql://postgres:postgres@localhost:5432/postgres",
    echo=True
)

# Создаем таблицы
Base.metadata.create_all(engine)

session_factory = sessionmaker(bind=engine)
```

```python
with session_factory() as session:
    users_data = [
        {"username": "Boris Muratshin", "email": 'muratshinboris@gmail.com'},
        {"username": "Alexey Kuzmin",   "email": 'lehakuz@gmail.com'},
        {"username": "Alina Trinz",     "email": 'coolalina@gmail.com'},
        {"username": "Ekaterina Nixon", "email": 'nixonkat@gmail.com'},
        {"username": "Donald Trump",    "email": 'realdonald@gmail.com'}
    ]

    users = []
    for user_data in users_data:
        user = User(username=user_data["username"], email=user_data["email"])
        users.append(user)
        session.add(user)
        print(f"Создан пользователь: {user.username}")

    session.commit()
    addresses_data = [
        {"street": "Michurina 43a", "city": "Yekaterinburg", "country": "Russia"},
        {"street": "Lenina 81",     "city": "Yekaterinburg", "country": "Russia"},
        {"street": "Taticheva 50",  "city": "Yekaterinburg", "country": "Russia"},
        {"street": "Tokarey 1",     "city": "Yekaterinburg", "country": "Russia"},
        {"street": "Gorykogo 4b",   "city": "Yekaterinburg", "country": "Russia"}
    ]

    for i, address_data in enumerate(addresses_data):
        address = Address(
            user_id=users[i].id,
            street=address_data["street"],
            city=address_data["city"],
            state="",
            zip_code="",
            country=address_data["country"],
            is_primary=True
        )
        session.add(address)
        print(f"Создан адрес для {users[i].username}: {address_data['street']}")

    session.commit()
```
